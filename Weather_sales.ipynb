```python
# loading libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load file path
df = pd.read_csv("DATA.csv")
```

## Exploatory Data Analysis (EDA)

```python
# Checking for missing values 

missing_values = df.isnull().sum()
missing_values
```

```python
# Hsitogram for numerical feature.

df.hist(bins=30, figsize=(20,15))
plt.suptitle("Histogram for Numerical features")
plt.show()
```

```python
df["Item"].value_counts()
```

```python
ax1 = df.groupby(['Item'], as_index = False)['Total'].sum()
ax1
plt.figure(figsize=(15,8))
plt.bar(ax1['Item'],ax1['Total'], color = 'orange')
plt.xlabel("Item", fontdict = {'fontsize': 14})
plt.ylabel("Total sales", fontdict = {'fontsize': 14})
plt.title("Total sales of items", fontdict = {'fontsize': 16})
for i,v in enumerate(ax1['Total']):
    plt.text(i,v,str(v), ha = 'center')
```
